```python
5. Data Discretisation:

For data discretisation, we have used pd.cut() method to discre5se COVID-19 Deaths into bins like.
This approach simplified the continuous data and made it more compatible with machine-learning models
that perform better with categorical inputs. The Age Group category used for demographic analysis was
retained and we have further categorised it into broader categories such as ‘Young Adult’, ‘Thirties’,
‘Forties’, etc.
```

```python
import pandas as pd

# Load the dataset
file_path = '/Users/akansharajput/Desktop/dapm.csv'  
data = pd.read_csv(file_path)

# Discretize 'COVID-19 Deaths' into buckets
death_bins = [0, 10, 100, 500, 1000, 5000, data['COVID-19 Deaths'].max()]
death_labels = ['1-10', '11-100', '101-500', '501-1000', '1001-5000', '5000+']
data['COVID-19 Deaths Discretized'] = pd.cut(data['COVID-19 Deaths'], bins=death_bins, labels=death_labels, include_lowest=True)

# Define a function to map age ranges to categories
def map_age_group(age_group):
    if age_group == '0-24':
        return 'Child/Young Adult'
    elif age_group == '25-34':
        return 'Young Adult'
    elif age_group == '35-44':
        return 'Thirties'
    elif age_group == '45-54':
        return 'Forties'
    elif age_group == '55-64':
        return 'Fifties'
    elif age_group == '65-74':
        return 'Senior'
    elif age_group == '75-84':
        return 'Elderly'
    elif age_group == '85+':
        return 'Elderly 85+'
    else:
        return 'Unknown'

# Apply the function to the 'Age Group' column
data['Age Group Discretized'] = data['Age Group'].apply(map_age_group)

# Save the transformed dataset
data.to_csv('/Users/akansharajput/Desktop/dapm.csv', index=False)  # Replace with the desired save path

# Display the first few rows of the transformed data
print(data.head())
```

```python
K-Means clustering analysis:

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/Users/akansharajput/Desktop/dapm.csv')  # Replace with your local file path

# Selecting numerical features for clustering
# Assuming 'COVID-19 Deaths' is the primary feature for clustering
features = data[['COVID-19 Deaths']]

# Standardizing the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Determining the optimal number of clusters using the Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(scaled_features)
    wcss.append(kmeans.inertia_)

# Plotting the results onto a line graph to observe 'The elbow'
plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')  # Within cluster sum of squares
plt.show()

# Assuming the optimal number of clusters is 3 based on the Elbow Method
optimal_clusters = 3  # Replace with the number you find optimal
kmeans_optimal = KMeans(n_clusters=optimal_clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
cluster_labels = kmeans_optimal.fit_predict(scaled_features)

# Calculate silhouette score
silhouette_avg = silhouette_score(scaled_features, cluster_labels)
print(f'Silhouette Score for {optimal_clusters} clusters: {silhouette_avg}')

# Assigning cluster labels to the original data
data['Cluster Labels'] = cluster_labels

# Save the clustered data to a new CSV file
data.to_csv('/Users/akansharajput/Desktop/dapm.csv', index=False)  # Replace with your desired save path
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import DBSCAN
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/akansharajput/Desktop/dapm.csv'
data = pd.read_csv(file_path)

# Check the range of years in the dataset
print("Years in the dataset:", data['Year'].unique())

# Objective 3: Seasonal trends in COVID-19 transmission and mortality
# Selecting relevant columns
data_obj3 = data[['Year', 'Month', 'Diseases', 'COVID-19 Deaths']]

# Preprocessing: Scale numerical features and encode categorical features
preprocessor_obj3 = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['Year', 'Month', 'COVID-19 Deaths']),
        ('cat', OneHotEncoder(), ['Diseases'])
    ])

# Adjust DBSCAN parameters
dbscan_obj3 = DBSCAN(eps=0.7, min_samples=10)  # Adjust these values as needed

# Create and apply the pipeline
pipeline_obj3 = Pipeline(steps=[('preprocessor', preprocessor_obj3),
                                ('cluster', dbscan_obj3)])
pipeline_obj3.fit(data_obj3)

# Add the cluster labels to the dataset
data_obj3['Cluster'] = pipeline_obj3['cluster'].labels_

# Display the first few rows of the modified dataset
print(data_obj3.head())

# Visualize the clustering results
plt.figure(figsize=(10, 6))
plt.scatter(data_obj3['Month'], data_obj3['COVID-19 Deaths'], c=data_obj3['Cluster'])
plt.title('Clustering Results for Seasonal Trends in COVID-19')
plt.xlabel('Month')
plt.ylabel('COVID-19 Deaths')
plt.colorbar(label='Cluster')
plt.show()
```
